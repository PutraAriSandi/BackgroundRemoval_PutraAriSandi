## Import Library

```python
from rembg import remove
from PIL import Image
```

## Proses Remove

```python
import matplotlib.pyplot as me

input_path = 'apel.jpg'
output_path = 'newfoto.png'

# Buka gambar asli
input_image = Image.open(input_path)

# Tampilkan gambar asli sebelum penghapusan latar belakang
me.subplot(1, 2, 1)
me.imshow(input_image)
me.title('Original Image')

# Hapus latar belakang
output_image = remove(input_image)

# Tampilkan gambar setelah penghapusan latar belakang
me.subplot(1, 2, 2)
me.imshow(output_image)
me.title('Image with Removed Background')

# Simpan gambar yang telah dihapus latar belakang
output_image.save(output_path)

# Tampilkan plot
me.tight_layout()
me.show()
```
